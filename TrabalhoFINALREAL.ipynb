```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('D:\Jupyter Notebook\ARTIGO TÓPÍCOS EM ECONOMETRIA\DadosJuntos.xlsx')
df
```

```python
df1 = df.loc[df['Valor de Mercado'].str.contains('mil')]
df1['Valor de Mercado'] = df1['Valor de Mercado'].map(lambda x: x.rstrip('mil € mil.'))
df1['Valor de Mercado'] = df1['Valor de Mercado'].astype(float)
df1['Valor de Mercado'] = df1['Valor de Mercado'].values/1000
df1
```

```python
df2 = df.loc[~df['Valor de Mercado'].str.contains('mil')]
df2['Valor de Mercado'] = df2['Valor de Mercado'].map(lambda x: x.rstrip('mi. € mil.'))
df2['Valor de Mercado'] = df2['Valor de Mercado'].replace(',','.', regex=True).astype(float)
df2
```

```python
df3 = pd.merge(df1, df2, how = 'outer')
df3
```

```python
df3['Valor de Mercado'] = df3['Valor de Mercado'].values*1000


df3['Accurate per game'] = df3['Accurate per game'].str[:4].replace(',','.', regex=True).astype(float)
df3['Acc, long balls'] = df3['Acc, long balls'].str[:4].replace(',','.', regex=True).astype(float)
df3['Acc, chipped passes'] = df3['Acc, chipped passes'].str[:4].replace(',','.', regex=True).astype(float)
df3['Acc, crosses'] = df3['Acc, crosses'].str[:4].replace(',','.', regex=True).astype(float)
df3['Succ, dribbles'] = df3['Succ, dribbles'].str[:4].replace(',','.', regex=True).astype(float)
df3['Total duels won'] = df3['Total duels won'].str[:4].replace(',','.', regex=True).astype(float)
df3['Ground duels won'] = df3['Ground duels won'].str[:4].replace(',','.', regex=True).astype(float)
df3['Aerial duels won'] = df3['Aerial duels won'].str[:4].replace(',','.', regex=True).astype(float)
df3['Acc, opposition half'] = df3['Acc, opposition half'].str[:4].replace(',','.', regex=True).astype(float)
df3['Acc, own half'] = df3['Acc, own half'].str[:4].replace(',','.', regex=True).astype(float)

df3['Idade'] = df3['Idade'].astype(float)
df3['Big chances created'] = df3['Big chances created'].astype(float)
df3['Assists'] = df3['Assists'].astype(float)
df3['Penalty won'] = df3['Penalty won'].astype(float)
df3['Big chances missed'] = df3['Big chances missed'].astype(float)
df3['Scoring frequency (Min)'] = df3['Scoring frequency (Min)'].astype(float)
df3['Goals'] = df3['Goals'].astype(float)
df3['Idade'] = df3['Idade'].astype(float)
df3['Team of the week'] = df3['Team of the week'].astype(float)
df3['Minutes per game'] = df3['Minutes per game'].astype(float)
df3['Total played'] = df3['Total played'].astype(float)
df3['Started'] = df3['Started'].astype(float)

df3 = df3.set_index('Jogador')
```
